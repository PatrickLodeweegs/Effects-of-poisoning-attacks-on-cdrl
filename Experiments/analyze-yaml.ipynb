```python
def read_config(filename, clean=True):
    with open(filename, "r") as configfile:
        config = yaml.safe_load(configfile)
    display_markdown("### " + config["name"], raw=True)
    display_markdown( config["descr"], raw=True)
    # model = config["model"]
    # experiment_ids = config["process_ids"]
    if config["model"] == "cdt4rec":
        # if clean:
        #     experiment_ids.append("232217")
        worst = "264560"
        worst = "144579"
        worst = read_log(*glob(f"../cdtrec-measurements/*{worst}*.log"))
    if config["model"] == "ddpg":
        if clean:
            config["process_ids"].append("536121")
        worst = "264560"
    display_markdown( f"In total {len(config['process_ids'])} experiments are included.", raw=True)
    config["logs"] = []
    for eid in config["process_ids"]: 
        if config["model"] == "cdt4rec":
            config["logs"].append(read_log(*glob(f"../cdtrec-measurements/*{eid}*.log")))
        if config["model"] == "ddpg":
            config["logs"].append(read_log(*glob(f"../ddpg-logs/ddpg-*{eid}.log")))
    return config

config = read_config(filename_cdt4rec1)
```

```python
def read_configs(filenames, name=None, clean=True):
    configs = []
    sconf = {}
    for filename in filenames:
        with open(filename, "r") as configfile:
            configs.append(yaml.safe_load(configfile))
    display_markdown("### Loaded configs", raw=True)
    if name is None:
        sconf["name"] = configs[0]["name"]
    else:

        sconf["name"] = name
    
    for config in configs:
        display_markdown("#### " + config["name"], raw=True)
        display_markdown( config["descr"], raw=True)
        display_markdown( f"This file includes {len(config['process_ids'])} experiments.", raw=True)


    # Check if all configs are for the same model
    assert all(map(lambda x: x["model"] == configs[0]["model"], configs))
    sconf["model"] = configs[0]["model"]
    if sconf["model"] == "cdt4rec":
        causal_param = lambda x: x["hyper_params"]["causal"] == configs[0]["hyper_params"]["causal"]
        assert  all(map(causal_param, configs))
        if all(map(causal_param, configs)):
            # If causal add
            worst = "264560"
        else:
            worst = "144579"
        sconf["worst"] = read_log(*glob(f"../cdtrec-measurements/*{worst}*.log"))
    else: # DDPG
        pass # No worst available :(
        
    sconf["poison_rates"] = sorted(list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["poison_rates"], configs))))
    # rates = list(set(config["poison_rates"]))
    sconf["process_ids"] = list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["process_ids"], configs)))
    sconf["triggers"] = list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["triggers"], configs)))
    print(list(map(lambda x: x["triggers"], configs)))

    sconf["logs"] = []
    for eid in sconf["process_ids"]: 
        if sconf["model"] == "cdt4rec":
            sconf["logs"].append(read_log(*glob(f"../cdtrec-measurements/*{eid}*.log")))
        if sconf["model"] == "ddpg":
            sconf["logs"].append(read_log(*glob(f"../ddpg-logs/ddpg-*{eid}.log")))
    return sconf
# config = read_configs([filename_cdt4rec1, filename_cdt4rec2, filename_cdt4rec3])
# config = read_configs([filename_ncdt4rec1, filename_ncdt4rec2, filename_ncdt4rec3])
config = read_configs([filename_ddpg])
```

```python
def config_plot_training_ctr(config, showtable=True, ymax=0.75):

    colors = {0.0: '#1f77b4', 1.0: '#ff7f0e', 5.0: '#2ca02c', 10.0:'#d62728',
              20.0:'#9467bd', 30.0:'#8c564b', 40.0:'#e377c2',}
    large_table = []
    window = 1000
    for trigger in filter(lambda x: x != "clean", config["triggers"]):
        logs_trigger = sorted(list(filter(lambda x: x["trigger"] in (trigger, "clean"), config["logs"])), key=lambda x: x["poisoning rate"])
        fig, ax = plt.subplots()
        for log in logs_trigger:
            # print(log["id"], len(log["mean clean ctr"]), len(reduce_data(log["mean clean ctr"], 1000)[0]))
            plot_training_ctr(ax, log, window, ["poisoning rate"], colors=colors)
            data = []
            if log["poisoning rate"]:
                row = [trigger, f"${log['poisoning rate']}$"] + [f"${round(a,2)} \pm {round(b,2)}$" for a, b in zip(*reduce_data(log["mean clean ctr"], window))]
                # print(len(data))
                data.append(row)
            large_table.extend(data)
            
        ax.set_xlabel("Episode")
        ax.set_ylim(-0.05,ymax)
        ax.set_yticks(np.arange(0, ymax, .2))
        ax.set_xticks(np.arange(0, 100_001, 10_000), minor=True)
        xax = np.arange(0, 100_001, 20_000)
        # ax.set_xticks(xax, [fr'\rotatebox{{45}}{{\numprint{{{round(x)}}}}}' for x in xax])
        ax.set_xticks(xax, [fr'{x:,}'.replace(",", r"\,") for x in xax], rotation=45)
        # ax.set_xticks(range(len(total_poisoning_rates)), total_poisoning_rates, rotation=45)
        ax.set_ylabel("CTR")
        ax.grid()
        fig.tight_layout()
        fig.set_size_inches(linewidth*.45,1.8)
        # plt.xticks(rotation=r'\rotatebox{45}{%s}')
        plt.savefig(f"{'-'.join(config['name'].split(' '))}-{trigger}-ctr-during-training.pgf",bbox_inches="tight")
        # ax.legend()
    print(f'THe clean baseline is {next(filter(lambda x: x["trigger"]  == ("rewardInvId") and x["poisoning rate"] == 10.0, config["logs"]))["id"]}')        
    mean, std = reduce_data(next(filter(lambda x: x["trigger"]  in ("clean"), config["logs"]))["mean clean ctr"], window)
    data = ["Clean", "$0$"] + [f"${round(a,2)} \pm {round(b,2)}$" for a, b in zip(mean, std)]
    large_table.append(data)
    print(f"{round(mean[-1], 2)} $\pm$ {round(std[-1],2)}")
    if showtable:
        print(pd.DataFrame.from_records(large_table, columns=["Trigger", "Poison rate"] + list(range(10000, 100_001, 10_000))).to_latex(escape=False, index=False))

# for config in configs:
config_plot_training_ctr(config, showtable=False, ymax=1.05)
```

```python
def config_plot_pr_overview(config, ymax=1.05):
    fig, ax = plt.subplots()
    window = 1000
    # rates = list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["poison_rates"], configs)))
    rates = config["poison_rates"]
    logs = config["logs"]
    # logs = list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["logs"], configs)))
    # triggers = list(reduce(lambda x, y: set(x) | set(y), map(lambda x: x["triggers"], configs)))
    # print(f"{rates=} {logs=} {triggers=} {configs[0]['poison_rates']}")
    triggers = config["triggers"]
    rate2point = {y:x+1 for x,y in enumerate(rates)}
    rate2point[0] = 0
    # Get the baseline
    # worst_mean, worst_std = reduce_data(worst["mean clean ctr"], window)
    # ax.axhline(y=worst_mean[-1], color='y', linestyle='-', label="fully inverted")
    # ax.fill_between([0, 7], worst_mean[-1] + worst_std[-1],
    #                 worst_mean[-1] - worst_std[-1], color="y", alpha=.2)
    # print(f"The worst possible pollicy is approximately {round(worst_mean[-1],2)} $\pm$ {round(worst_std[-1], 2)}")
    
    # Get the experiments
    table = {}
    for trigger in filter(lambda x: x != "clean", triggers):
        print(f"Plotting {trigger}")
        logs_per_trigger = sorted(filter(lambda x: x["trigger"]  in (trigger, "clean"), logs), key=lambda x: x["poisoning rate"])
        print(f"Found {len(logs_per_trigger)} logs")
        points = []
        row = []
        for log in logs_per_trigger:
            mean, std = reduce_data(log["mean clean ctr"], window)
            index = rate2point.get(log["poisoning rate"], log["poisoning rate"])
            # print(f"Collecting data for {log['poisoning rate']}")
            # print("\t", (index, mean[-1], std[-1]))
            points.append((index, mean[-1], std[-1]))
            if log["poisoning rate"]:
                row.append(( f"{round(mean[-1],2)} $\pm$ {round(std[-1],2)}"))
        ax.errorbar(*zip(*points), label=trigger, linestyle="dotted",
                    marker=".", markersize=5, alpha=0.8, lw=1)
        # print(list(zip(*points))[0])
    
        table[trigger] = row
    ax.set_xlabel("Poisoning rate")
    # ax.set_ylim(-0.05,.85) # 
    ax.set_ylim(-0.05, ymax)
    ax.grid()
    print(list(range(len(rates))), rates)
    print(rate2point)
    ticks = [0] + list(map(lambda x: x+1, next(zip(*enumerate(rates)))))
    print(ticks)
    ax.set_xticks(ticks, ["clean"] + [fr"{round(r)}\%" for r in rates], rotation=45)
    ax.set_yticks(np.arange(0, ymax, .2))
    ax.set_ylabel("CTR")
    ax.legend(prop={'size': small},loc='right', bbox_to_anchor=(1.75, 0.5),)
    # fig.tight_layout()
    fig.set_size_inches(linewidth-2,2.)
    name = f"{'-'.join(config['name'].split(' '))}-plot.pgf"
    print(f"Saving to: {name}")
    plt.savefig(name, bbox_inches="tight",  backend="pgf")

config_plot_pr_overview(config, 1.05)
```

```python
def config_plot_relative_effect(config, ):
    window = 1000
    rates = [f"{round(pr)}%" for pr in config["poison_rates"]]
    logs = config["logs"]
    triggers = config["triggers"]
    rate2point = {y:x+1 for x,y in enumerate(rates)}
    rate2point[0] = 0
    
    table = {}
    clean = next(filter(lambda x: x["trigger"]  == "clean", logs))
    meanc, stdc = reduce_data(clean["mean clean ctr"], window)
    ref = meanc[-1]
    print("Clean: ", )
    for trigger in filter(lambda x: x != "clean", triggers):
        print(f"Eval {trigger}")
        logs_per_trigger = sorted(filter(lambda x: x["trigger"]  == trigger, logs), key=lambda x: x["poisoning rate"])
        print(f"\tFound {len(logs_per_trigger)} logs")
        points = []
        row = []
        for log in logs_per_trigger:
            mean, std = reduce_data(log["mean clean ctr"], window)
            index = rate2point.get(log["poisoning rate"], log["poisoning rate"])

            row.append(( f"{round(mean[-1]/ref * 100,2)}%"))
        if len(row) < len(rates):
            row.extend(["-" for _ in range(len(rates)-len(row))])
        table[trigger] = row

    print(list(range(len(rates))), rates)
    print(rate2point)
    return (pd.DataFrame(table, index=rates))
df = config_plot_relative_effect(config)
print(df.to_latex())
```
